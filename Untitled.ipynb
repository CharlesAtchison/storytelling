# Storytelling Project

### Goal
* Use data analysis to reduce customer attrition/churn for customers within the month to month plan.

```python
telco_df = pd.read_csv('telco_csv_data.csv')
mtm = telco_df[telco_df.contract_type == 0] # This is the month-to-month plan under contract type
mtm.drop(columns='contract_type', inplace=True)
total_mtm_representation_pct = (len(mtm)/len(telco_df)) * 100
mtm.describe()
```

### Inital Observations for MTM contracts (month-to-month: MTM)

* 20.82% of them are senior citizens.
* Mean tenure is 18 months.
* Mean monthly charges are about $66.00 a month.
* There are 3,876 people who have elected the MTM, contract representing about 55% of the data
